```python
# MedLang: Simple Medical Diagnosis System

# Step 1: Define medical rules
rules = {
    "Flu": ["fever", "cough", "headache"],
    "Cold": ["cough", "sneezing"],
    "Malaria": ["fever", "chills", "sweating"]
}

# Step 2: Take symptoms from user
user_input = input("Enter symptoms separated by comma: ")
user_symptoms = user_input.lower().split(",")

# Remove extra spaces
user_symptoms = [s.strip() for s in user_symptoms]

# Step 3: Diagnosis engine
possible_diseases = []

for disease, symptoms in rules.items():
    if all(symptom in user_symptoms for symptom in symptoms):
        possible_diseases.append(disease)

# Step 4: Output result
if possible_diseases:
    print("Possible Diagnosis:")
    for d in possible_diseases:
        print("-", d)
else:
    print("No matching disease found. Please consult a doctor.")


```
